## Visualizations of Data

```python
daily_activity.head()
```

```python
numeric_columns = daily_activity.select_dtypes(include=['number'])
```

```python
correlation_matrix = numeric_columns.corr()
```

```python
plt.figure(figsize=(14, 14))
sns.heatmap(correlation_matrix, cmap='Reds', annot=True)
plt.show()
```

```python
daily_activity.hist(figsize=(20,20),color = 'Olive')
```

```python
da_days = daily_activity.groupby('Day_of_week').agg({'TotalDistance':'sum'}).reset_index().sort_values('TotalDistance',ascending = False)
a1 = (8, 6)
fig, ax = plt.subplots(figsize=a1)
color_palette = ["green", "orange", "yellow", "blue", "red", "purple", "Peru"]
plot= sns.barplot(x="Day_of_week", y="TotalDistance", data=da_days,palette =color_palette)
plot.set_xticklabels(ax.get_xticklabels(),rotation=90)
plot.set_title('Total Distance During Each Day')
ax.set_ylabel('Total Distance Sum')
ax.set_xlabel('Days')
```

After May 5, 2016 the calories started to decrease, and after May 9 it's decreasing continuously

```python
all_calories = daily_activity.groupby('ActivityDate', as_index=False)['Calories'].sum()

daily_calories = go.Scatter(x=all_calories['ActivityDate'], y=all_calories['Calories'], line=dict(color='Red'))
layout = go.Layout(title='Daily Calories', xaxis=dict(title='Date'), yaxis=dict(title='Calories'), plot_bgcolor='white')
fig = go.Figure(data=[daily_calories], layout=layout)
iplot(fig)
```

```python
all_sleep = sleep_per_day.groupby('Date', as_index=False)['TotalMinutesAsleep'].sum()

daily_sleep = go.Scatter(x=all_sleep['Date'], y=all_sleep['TotalMinutesAsleep'], line=dict(color='Red'))
layout = go.Layout(title='Total Minutes As Sleep Each Day', xaxis=dict(title='Date'), yaxis=dict(title='Minutes As Sleep'),plot_bgcolor='white')
fig = go.Figure(data=[daily_sleep], layout=layout)
iplot(fig)
```

```python
spd_days = sleep_per_day.groupby('Day_of_week').agg({'TotalMinutesAsleep':'sum'}).reset_index().sort_values('TotalMinutesAsleep',ascending = False)
a1 = (8, 6)
fig, ax = plt.subplots(figsize=a1)
color_palette = ["green", "orange", "yellow", "blue", "red", "purple", "Peru"]
plot= sns.barplot(x="Day_of_week", y="TotalMinutesAsleep", data=spd_days,palette =color_palette)
plot.set_xticklabels(ax.get_xticklabels(),rotation=90)
plot.set_title('Total Sleep Minutes During Each Day')
ax.set_ylabel('Total Sleep Minutes Sum')
ax.set_xlabel('Days')
```

```python
spd_days = sleep_per_day.groupby('Month').agg({'TotalMinutesAsleep':'sum'}).reset_index().sort_values('TotalMinutesAsleep',ascending = False)

a1 = (6, 8)
fig, ax = plt.subplots(figsize=a1)
color_palette = ["green", "orange", "yellow", "blue", "red", "purple", "Peru"]
plot= sns.barplot(x="Month", y="TotalMinutesAsleep", data=spd_days,palette =color_palette)
plot.set_xticklabels(ax.get_xticklabels(),rotation=90)
plot.set_title('Total Sleep Minutes During Month 4 and 5')
ax.set_ylabel('Total Sleep Minutes Sum')
ax.set_xlabel('Months')
```
